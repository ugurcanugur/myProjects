```python
import numpy
""" seed here seferinde aynı rand saglıyo"""
numpy.random.seed(12)
T = 300

###### INITIALIZATION ######
Q = 0 
B0 = 0
B1 = 0
B2 = 0 

futureEventsList = []
futureEventsList.append([0, "Arrival"])
futureEventsList.append([T, "End"])

int_Q = 0
int_B0 = 0
int_B1 = 0
int_B2 = 0

t_prev = 0 

Arrival_Server = ["-","-","-"] ## Let's create a list with 3 positions corresponding to 3 servers
Arrival_Queue = []

P = 0
sum_TS = 0

N = 0
sum_WQ = 0

Q_max = 0
TS_max = 0
WQ_max = 0


while(len(futureEventsList) > 0):
  futureEventsList.sort()
  event = futureEventsList.pop(0)
  t = event[0]
  """ q ve b toplam hesapları """
  int_Q = int_Q + (t - t_prev) * Q
  int_B0 = int_B0 + (t - t_prev) * B0
  int_B1 = int_B1 + (t - t_prev) * B1
  int_B2 = int_B2 + (t - t_prev) * B2


  if event[1] == "Arrival":
    ia = numpy.random.exponential(scale = 1/2)
    futureEventsList.append([t + ia, "Arrival"])
    """ makinaların tamamı doluysa que ekleme """
    if B0 + B1 + B2 == 3:
      Q = Q + 1
      Arrival_Queue.append(t)

      Q_max = max(Q_max , Q)
      """ sisteme giricek urune procces time belirleme """
    else:
      N = N+1
      s = numpy.random.randint(3,6)

## this version of the code assumes if drill 1 is idle, the part goes to drill 1
## if drill 1 is busy and drill 2 is idle, the part goes to drill 2
## if drill 1 and 2 are both busy, the part goes to drill 3
## other selection rules are also possible
      """s rand procces time add to general time and find the departure time """

      if B0 == 0:
        B0 = 1
        futureEventsList.append([t+s, "Departure", 0]) ## the last element of Departure event shows which drill the part will depart from
        Arrival_Server[0] = t ## instead of appending and changing the size of the list, we will place the element in the corresponding position


      elif B1 == 0:
        B1 = 1
        futureEventsList.append([t+s, "Departure", 1]) ## the last element of Departure event shows which drill the part will depart from
        Arrival_Server[1] = t ## instead of appending and changing the size of the list, we will place the element in the corresponding position


      else:
        B2 = 1
        futureEventsList.append([t+s, "Departure", 2]) ## the last element of Departure event shows which drill the part will depart from
        Arrival_Server[2] = t ## instead of appending and changing the size of the list, we place the element in the corresponding position




  elif event[1] == "Departure":

    machineID = event[2] ## get the information on which drill the part is departing from
    """sistemde ne kadar zaman gecirdiğini bulma """
    P = P + 1
    X = Arrival_Server[machineID] ## instead of popping and changing the size of the list, we take the value of the element in the corresponding position
    sum_TS = sum_TS + (t - X)
    Arrival_Server[machineID] = "-"

    TS_max = max(TS_max , t - X)
    """depature etme makineyi 0 lama  """
    if Q == 0:
      if machineID == 0:
        B0 = 0
      elif machineID == 1:
        B1 = 0
      else:
        B2 = 0
      """makina ve que doluyken departure etme   """
    else:
      Q = Q-1
      s = numpy.random.randint(3,6)
      futureEventsList.append([t+s, "Departure", machineID]) #schedule departure from the same machine

      X = Arrival_Queue.pop(0)
      Arrival_Server[machineID] = X

      N = N+1
      sum_WQ = sum_WQ + (t - X)
      WQ_max = max(WQ_max , t - X )


  elif event[1] == "End":
    futureEventsList.clear()

# Optional - if you want to trace the code
#  print(event, Q, B0, B1,B2, Arrival_Queue, Arrival_Server)
  t_prev = t


print("Average Queue Length = ", int_Q/T)
print("Maximum Queue Length = ", Q_max)
print("Average Utilization of Drill 1 = ", int_B0/T)
print("Average Utilization of Drill 2 = ", int_B1/T)
print("Average Utilization of Drill 3 = ", int_B2/T)
print("Average Total Time in System = ", sum_TS/P)
print("Max Total Time in System = ", TS_max)
print("Average Waiting Time in Queue = ", sum_WQ/N)
print("Max Waiting Time in Queue = ", WQ_max)
```
